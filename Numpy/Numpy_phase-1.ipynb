```python
import numpy as np
```

## Numpy Array

```python
array_1d = np.array([[1,2,3,4,5]])
print("1D array : ",array_1d)
array_2d = np.array([[1,2,3],[4,5,6]])
print("2D array : ",array_2d)
```

## Creating Arrays

```python
zero = np.zeros((3,4))
print("Zeros array::\n",zero)

one = np.ones((3,4))
print("Ones Array::\n",one)

fulls = np.full((3,4),7)
print("Full Matrix Array::\n",fulls)

randoms = np.random.random((2,3))
print("Random Matrix Array::\n",randoms)

secquence = np.arange(0,10,2)
print("Secquence Matrix Array::\n",secquence)
```

## Vector , Matrix , Tensors

```python
vector  = np.array([1,4,6,2,6,7])
print("Vector::\n",vector)

matrix = np.array([[1,2,3],
                 [4,5,6]])
print("Matrix::\n",matrix)


tensor = np.array([[[1,2],[3,4]],
                    [[4,5],[6,7]]])
print("tensor::\n",tensor)

```

## Array Properties

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print("Shape::",arr.shape)
print("Dimentions::",arr.ndim)
print("Size::",arr.size)
print("Data Type::",arr.dtype)
```

## Array Reshaping

```python
arr = np.arange(12)
print("Original Array::",arr)

reshaped = arr.reshape(3,4)
print("Reshaped Array::",reshaped)

flatened = reshaped.flatten()
print("Flattened Array::",flatened)

raveled = reshaped.ravel()
print("Raveled Array::",raveled)

transposed = reshaped.T
print("Transposed Array::",transposed)
```
