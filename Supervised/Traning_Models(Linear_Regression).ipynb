## Equation 

y^ = 0o+01x1+02x2+0nxn 

y = predicted value
n = number of features 
xi = ith feature of value 
0j = jth parameter (including the bias term 0o)
0o = bais term

Vector Form

y^ = h0(x)=0.x


## MSE of Linear Regression

MSE(X,h0) = 1/m sum (0T x^i -y^i)^2

## The Normal Equation

0^ = (X^T X)^-1 X^T y

## Generating Linear Looking data

```python
import numpy as np 
```

```python
X = 2*np.random.rand(100,1)
y = 4 + 3 * X + np.random.rand(100,1)
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y,"b.")
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)),X]  # add x0 = 1 to each instance
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_best)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)  # True relation: y = 2x

# Initialize parameters
m = 0.0   # slope
b = 0.0   # intercept
alpha = 0.05  # learning rate
epochs = 40  # fewer iterations so we can visualize each step

# Plot setup
plt.figure(figsize=(10,6))
plt.scatter(X, y, color="blue", label="Data points")

# Gradient Descent with visualization
for i in range(epochs):
    y_pred = m * X + b
    error = y - y_pred
    
    # Gradients
    dm = -(2/len(X)) * np.sum(X * error)
    db = -(2/len(X)) * np.sum(error)
    
    # Update parameters
    m -= alpha * dm
    b -= alpha * db
    
    # Plot regression line at selected iterations
    if i % 5 == 0:  # plot every 5 iterations
        plt.plot(X, m*X + b, label=f"Iteration {i}")

plt.title("Regression Line Improving Each Iteration")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()



print(f"Final slope (m): {m:.4f}")
print(f"Final intercept (b): {b:.4f}")

# 5. Visualization
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, m*X + b, color="red", label="Regression line")
plt.title("Linear Regression with Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
