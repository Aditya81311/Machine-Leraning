## Equation 

y^ = 0o+01x1+02x2+0nxn 

y = predicted value
n = number of features 
xi = ith feature of value 
0j = jth parameter (including the bias term 0o)
0o = bais term

Vector Form

y^ = h0(x)=0.x


## MSE of Linear Regression

MSE(X,h0) = 1/m sum (0T x^i -y^i)^2

## The Normal Equation

0^ = (X^T X)^-1 X^T y

## Generating Linear Looking data

```python
import numpy as np 
```

```python
X = 2*np.random.rand(100,1)
y = 4 + 3 * X + np.random.rand(100,1)
```

```python
import matplotlib.pyplot as plt
plt.plot(X,y,"b.")
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)),X]  # add x0 = 1 to each instance
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_best)
```

```python
y_predict
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X,y)
print(lin_reg.intercept_)
print(lin_reg.coef_)
print(lin_reg.predict(X_new))
```

```python
pinverse = np.linalg.pinv(X_b).dot(y)
```

```python
pred = X_new_b.dot(pinverse)
pred
```
