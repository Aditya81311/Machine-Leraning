```python
from sklearn.datasets import fetch_openml
import pandas as pd 
import numpy as np 
mnist = fetch_openml('mnist_784')

X = mnist["data"]
y = mnist["target"]

print(f"{X.shape}\n{y.shape}")
```

```python
y = y.astype(np.uint8)
```

```python
X_train,X_test = X[:60000],X[60000:]
y_train,y_test =  y[:60000],y[60000:]
```

```python
from sklearn.linear_model import  SGDClassifier
```

```python
y_train_3 = (y_train == 3)
y_test_3 = (y_test == 3)
sgd_clf = SGDClassifier(random_state = 42)
sgd_model = sgd_clf.fit(X_train,y_train_3)
```

```python
from sklearn.model_selection import cross_val_predict
some_digit = X.iloc[0]
some_digit_image = some_digit.to_numpy().reshape(28,28)
y_pred = sgd_model.predict([some_digit])
y_pred
```

```python
import matplotlib.pyplot as plt
plt.imshow(some_digit_image,cmap = "binary")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(sgd_clf,X_train,y_train_3,cv = 3, scoring = "accuracy")
scores
```

```python
from sklearn.metrics import confusion_matrix ,f1_score
from sklearn.model_selection import cross_val_predict
val_scores = cross_val_predict(sgd_clf,X_train,y_train_3,cv = 3)
```

```python
sgd_scores  = confusion_matrix(y_train_3,val_scores)
sgd_scores
```

```python
from sklearn.neighbors import KNeighborsClassifier
k_clf = KNeighborsClassifier()
k_model = k_clf.fit(X_train,y_train_3)
```

```python
k_pred = k_model.predict([some_digit])
k_pred
```

```python
k_val_scores = cross_val_predict(k_clf,X_train,y_train_3,cv = 3)
print(k_val_scores)
k_mat  = confusion_matrix(y_train_3,k_val_scores)
print(k_mat)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [{'weights':['distance','uniform'],'n_neighbors':[3, 5, 7, 9, 11], 'metric': ['minkowski', 'euclidean', 'manhattan'],
    'p': [1, 2]}]
grid_search = GridSearchCV(  estimator=k_clf,param_grid,cv = 5,scoring = "accuracy", n_jobs=-1 )
grid_search.fit(X_train,y_train_3)
```
